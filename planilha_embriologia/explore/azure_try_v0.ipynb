```python
from office365.runtime.auth.authentication_context import AuthenticationContext
from office365.sharepoint.client_context import ClientContext
from office365.sharepoint.files.file import File
import io
import pandas as pd

# SharePoint site and file info
site_url = 'https://huntington.sharepoint.com/' 
relative_file_url = "sites/WorkDados/EMBRIOLOGIA/PLANILHA DE CASOS/2025/CASOS 2025 IBI.xlsx"  # Path to file in SharePoint

# Azure AD App credentials
client_id = "6df26b35-48da-49cf-8733-5c87d22bc6fb"
client_secret = "de91fe84-7954-485a-a99a-1ad335d29328"
tenant_id = "4055bf2a-b04a-40e4-9f2a-50000c55545b"

# Authenticate
app_principal = {
    'client_id': client_id,
    'client_secret': client_secret,
}

ctx_auth = AuthenticationContext(site_url)
if ctx_auth.acquire_token_for_app(client_id=client_id, client_secret=client_secret):
    ctx = ClientContext(site_url, ctx_auth)
    
    # Download file as bytes
    response = File.open_binary(ctx, relative_file_url)
    
    if response.status_code == 200:
        # Load into pandas
        bytes_file_obj = io.BytesIO(response.content)
        df = pd.read_excel(bytes_file_obj, engine='openpyxl')
        print(df.head())
    else:
        print(f"Failed to download file: {response.status_code}")
else:
    print("Authentication failed")
```
